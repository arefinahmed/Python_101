```python
game=[[0,0,0],[0,0,0],[0,0,0]]

 


    
def game_board():
    print('   1',' 2', ' 3') 
    for row,column in enumerate(game):
        print(row, column)
        row+=1


game_board() # showing the game board

print("Enter the first player turn")

game[0][0]=1

print('\n')

game_board()
```
